Lets change the localization distance to 5, 10, 25, 50, 100, 500 and 1000

```python
import matplotlib.pyplot as plt
import numpy as np
import true_state
import tmp.enkf25_loc_l5_results as enkf25_loc_l5
import tmp.enkf25_loc_l10_results as enkf25_loc_l10
import tmp.enkf25_loc_l25_results as enkf25_loc_l25
import tmp.enkf25_loc_l50_results as enkf25_loc_l50
import tmp.enkf25_loc_l100_results as enkf25_loc_l100
import tmp.enkf25_loc_l500_results as enkf25_loc_l500
import tmp.enkf25_loc_l1000_results as enkf25_loc_l1000
```

```python
dist=[5, 10, 25, 50, 100, 500, 1000]
d_a=7*[None]
d_b=7*[None]
d_a[0]=np.linalg.norm(true_state.a_true_300-enkf25_loc_l5.x_a[60][0:1000])
d_a[1]=np.linalg.norm(true_state.a_true_300-enkf25_loc_l10.x_a[60][0:1000])
d_a[2]=np.linalg.norm(true_state.a_true_300-enkf25_loc_l25.x_a[60][0:1000])
d_a[3]=np.linalg.norm(true_state.a_true_300-enkf25_loc_l50.x_a[60][0:1000])
d_a[4]=np.linalg.norm(true_state.a_true_300-enkf25_loc_l100.x_a[60][0:1000])
d_a[5]=np.linalg.norm(true_state.a_true_300-enkf25_loc_l500.x_a[60][0:1000])
d_a[6]=np.linalg.norm(true_state.a_true_300-enkf25_loc_l1000.x_a[60][0:1000])
d_b[0]=np.linalg.norm(true_state.b_true_300-enkf25_loc_l5.x_a[60][1000:2000])
d_b[1]=np.linalg.norm(true_state.b_true_300-enkf25_loc_l10.x_a[60][1000:2000])
d_b[2]=np.linalg.norm(true_state.b_true_300-enkf25_loc_l25.x_a[60][1000:2000])
d_b[3]=np.linalg.norm(true_state.b_true_300-enkf25_loc_l50.x_a[60][1000:2000])
d_b[4]=np.linalg.norm(true_state.b_true_300-enkf25_loc_l100.x_a[60][1000:2000])
d_b[5]=np.linalg.norm(true_state.b_true_300-enkf25_loc_l500.x_a[60][1000:2000])
d_b[6]=np.linalg.norm(true_state.b_true_300-enkf25_loc_l1000.x_a[60][1000:2000])
```

```python
imbal=7*[None]
from check_balance import compute_b
imbal[0]=np.linalg.norm(compute_b(enkf25_loc_l5.x_a[60][0:1000])-enkf25_loc_l5.x_a[60][1000:2000])
imbal[1]=np.linalg.norm(compute_b(enkf25_loc_l10.x_a[60][0:1000])-enkf25_loc_l10.x_a[60][1000:2000])
imbal[2]=np.linalg.norm(compute_b(enkf25_loc_l25.x_a[60][0:1000])-enkf25_loc_l25.x_a[60][1000:2000])
imbal[3]=np.linalg.norm(compute_b(enkf25_loc_l50.x_a[60][0:1000])-enkf25_loc_l50.x_a[60][1000:2000])
imbal[4]=np.linalg.norm(compute_b(enkf25_loc_l100.x_a[60][0:1000])-enkf25_loc_l100.x_a[60][1000:2000])
imbal[5]=np.linalg.norm(compute_b(enkf25_loc_l500.x_a[60][0:1000])-enkf25_loc_l500.x_a[60][1000:2000])
imbal[6]=np.linalg.norm(compute_b(enkf25_loc_l1000.x_a[60][0:1000])-enkf25_loc_l1000.x_a[60][1000:2000])
```

```python
plt.plot(dist,d_a)
plt.plot(dist,d_b)
plt.plot(dist,imbal)
plt.legend(["$|true_b-x_b|_2$","$|true_b-x_b|_2$","$|$imbalance variable b$|_2$"])
```
